```python
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv(r"../data/processed/after_eda_ecommerce_data.csv")
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.figsize']=(10, 6)
matplotlib.rcParams['figure.facecolor']='#00000000'
%matplotlib inline
```

```python
fig = px.histogram(df, 
                   x='Gender', 
                   marginal='box', 
                   color='Discount Applied', 
                   color_discrete_sequence=['green', 'grey'])
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig=px.histogram(df, x="Age")
fig.update_layout(bargap=0.1)
fig.show()
```

```python
fig=px.scatter(df, 
               y='Total Spend', 
               x='Days Since Last Purchase', 
               color='Gender',  
               color_discrete_sequence=['red', 'blue'],          
               opacity=0.8)
fig.show()
```

```python
fig=px.scatter(df, 
               y='Total Spend', 
               x='Items Purchased', 
               color='Gender',  
               color_discrete_sequence=['red', 'blue'],          
               opacity=0.8)
fig.show()
```

```python
sns.boxplot(x='Satisfaction Level', y='Age', data=df)
```
